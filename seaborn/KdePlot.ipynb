```python
import pandas as pd
```

```python
df=pd.read_csv('auto-mpg.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.horsepower.replace('?',0,inplace=True)
```

```python
df.info()
```

```python
df.horsepower=df.horsepower.astype(float)
```

```python
sns.kdeplot(df.horsepower)
```

```python
sns.kdeplot(df.horsepower,shade=True)
```

```python
sns.kdeplot(df.horsepower,bw=.1)
```

```python
sns.kdeplot(df.horsepower,cumulative=True)
```

```python
sns.kdeplot(df.horsepower,df.mpg,shade=True)
```

```python
sns.kdeplot(df.horsepower,df.mpg,n_levels=15)
```

```python
sns.kdeplot(df.horsepower,df.mpg,shade=True,cbar=True)
```
