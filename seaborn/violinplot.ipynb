```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
df.head()
```

```python
df.cylinders.value_counts()
```

```python
df=df[df.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(df.displacement)
```

```python
df.displacement.value_counts()
```

```python
sns.set_style("whitegrid")
```

```python
sns.violinplot(x=df.cylinders,y=df.displacement)
```

```python
sns.violinplot(x=df.cylinders,y=df.displacement,hue='origin',data=df)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=df[df.origin.isin(['japan','europe'])])
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=df[df.origin.isin(['japan','europe'])],split=True)
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=df[df.origin.isin(['japan','europe'])],split=True,inner='quartiles')
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=df[df.origin.isin(['japan','europe'])],split=True,inner='quartiles',
               scale='count')
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=df[df.origin.isin(['japan','europe'])],split=True,inner='quartiles',
               scale='count',scale_hue=False)
plt.legend(loc=2)
```
