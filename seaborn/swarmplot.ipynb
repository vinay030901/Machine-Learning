```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('mpg').dropna()
```

```python
sns.set_style('whitegrid')
```

```python
df.cylinders.value_counts()
```

```python
df=df[df.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(df.horsepower)
```

```python
sns.swarmplot(x=df.origin,y=df.horsepower)
```

```python
sns.swarmplot(x=df.origin,y=df.horsepower,hue=df.cylinders)
```

```python
sns.swarmplot(x=df.origin,y=df.horsepower,hue=df.cylinders,dodge=True)
```
