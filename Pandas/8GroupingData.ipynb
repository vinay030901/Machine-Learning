```python
country_use_python=cnt_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_respondents=df.Country.value_counts()
country_respondents
```

```python
python_df=pd.concat([country_respondents,country_use_python],axis='columns',sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondents','LanguageWorkedWith':'NumKnowsPython'},inplace=True)
python_df
```

```python
python_df['PctKnowsPython']=(python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df
```
